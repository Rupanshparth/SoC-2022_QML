```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_bloch_multivector, plot_histogram
sim = Aer.get_backend('aer_simulator')
import numpy as np
from numpy import pi
```

```python
n = 5
pe = QuantumCircuit(n+1,n)
pe.x(n)
a = 1
for i in range(n):
    pe.h(i)
for i in range(n):
    for q in range(a):
        pe.cp(pi/4,i,n)
    a *= 2
    
pe.barrier()

for qubit in range(int(n/2)):
    pe.swap(qubit, n-qubit-1)
for j in range(n):
    for m in range(j):
        pe.cp(-(pi)/float(2**(j-m)), m, j)
    pe.h(j) 

pe.barrier()  

for i in range(n):
    pe.measure(i,i)

pe.draw()

    
```
